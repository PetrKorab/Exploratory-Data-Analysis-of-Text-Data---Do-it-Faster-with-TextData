```python
import text_data
import pandas as pd
import string
```

```python
data = pd.read_csv('train.csv', encoding='utf8')
data = data.head(n=5000)
data = pd.DataFrame(data)
data
```

```python
# Data cleaning

# stopwords removal
from nltk.corpus import stopwords
import re

stopwords = set(stopwords.words("english"))

print(stopwords)

def remove_stopwords(text):
    filtered_words = [word.lower() for word in text.split() if word.lower() not in stopwords]
    return " ".join(filtered_words)

# removing numbers and punctuations

def preprocess(text):
    text_input = re.sub(r'!', ' ', str(text))
    text_input = re.sub(r'[0-9]+', '', str(text_input))
    text_input = text_input.replace('<br /><br />', '')
    text_input = text_input.replace('(', '')
    text_input = text_input.replace('"', '')
    text_input = text_input.replace('$', '')
    text_input = text_input.replace('*', '')
    text_input = text_input.replace('.', '')
    text_input = text_input.replace('/', '')
    text_input = text_input.replace(':', '')
    text_input = text_input.replace('?', '')
    text_input = text_input.replace(',', '')
    text_input = text_input.replace('---', '')
    text_input = text_input.replace("""'""", '')
    output = text_input.replace(')', '')
    return output.lower().strip()


data['text'] = data.text.map(remove_stopwords)
data['text'] = data.text.map(preprocess)

data_filtered=data['text']
```

```python
data_filtered.to_csv('data_cleaned.csv', encoding = 'utf8')
```

```python
# creating the corpus

corpus = text_data.Corpus(list(data_filtered))
```

```python
pos_sentiment = data[data.sentiment == "pos"]
neg_sentiment = data[data.sentiment == "neg"]
```

```python
pos_sentiment_corpus = corpus.slice(set(pos_sentiment.index))
neg_sentiment_corpus = corpus.slice(set(neg_sentiment.index))
```

```python
# printing 10 most frequent words in the corpus with positive label
pos_sentiment_corpus.most_common(10)
```

```python
# printing 10 most frequent words in the corpus with negative label
neg_sentiment_corpus.most_common(10)
```

```python
# most frequent bi-grams - positive sentiment
pos_sentiment_corpus.add_ngram_index(n=2)
bigrams_positive = pos_sentiment_corpus.ngram_indexes[2].most_common(10)
bigrams_positive
```

```python
# bar chart of top bigrams Frequency - positive sentiment
import numpy as np

text_data.display.render_bar_chart(labels = np.array(bigrams_positive)[:,0],
                                   vector_data = np.array(bigrams_positive)[:,1],
                                   x_label='Frequency',
                                   y_label='Word')
```

```python
# most frequent tri-grams - positive sentiment
pos_sentiment_corpus.add_ngram_index(n=3)
trigrams_positive = pos_sentiment_corpus.ngram_indexes[3].most_common(10)
trigrams_positive
```

```python
# bar chart of top trigrams frequency - positive sentiment
import numpy as np

array = np.array(trigrams_positive)
text_data.display.render_bar_chart(labels = array[:,0],
                                   vector_data = array[:,1], x_label='Frequency', y_label='Word')
```
